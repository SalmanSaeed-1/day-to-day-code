```python
def pair_sum(t, n):
    result = []

    for i in range(len(t)):
        for j in range(i + 1, len(t)):
            if t[i] + t[j] == n:
                pair = (min(t[i], t[j]), max(t[i], t[j]))
                if pair not in result:
                    result.append(pair)

    return tuple(result)


numbers = (1, 3, 5, 7, 9, 2, 4, 6, 8)
print(pair_sum(numbers, 10))
```

```python
def tuple_transform(t):
    result = []

    for i in t:
        if len(i) % 2 == 0:
            result.append(i.upper())
        else:
            result.append(i[::-1].lower())

    return tuple(result)


words = ("apple", "banana", "kiwi", "orange", "mango")
print(tuple_transform(words))
```
